```python
import pickle
import matplotlib.pyplot as plt
from ex_advanced import moving_average_curve
```

```python
with open('history/0016_history.dump', 'rb') as f:
  history = pickle.load(f)
```

```python
figure = plt.figure()
plt.plot(history['loss'], label='loss')
if 'val_loss' in history:
  plt.plot(moving_average_curve(history['val_loss'], 5), label='val_loss')
plt.legend()
plt.ylim((0, 0.15))
plt.show()
```

```python
figure = plt.figure()
plt.plot(history['acc'], label='acc')
if 'val_acc' in history:
  plt.plot(moving_average_curve(history['val_acc'], 5), label='val_acc')
plt.legend()
plt.ylim((0.97, 1))
plt.show()
```
